```python
import time
import os
import urllib.request
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup as bs
from fake_useragent import UserAgent

# Create the 'top' directory if it doesn't exist
os.makedirs('./top', exist_ok=True)

# Configure Chrome options
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--ignore-certificate-errors')
chrome_options.add_argument('--ignore-ssl-errors')
chrome_options.add_argument('--disable-blink-features=AutomationControlled')
chrome_options.add_argument('--disable-popup-blocking')
chrome_options.add_argument('--disable-notifications')

# Set a custom User-Agent header using fake-useragent
ua = UserAgent()
user_agent = ua.random
chrome_options.add_argument(f'user-agent={user_agent}')

# Initialize Chrome WebDriver with configured options
driver = webdriver.Chrome(options=chrome_options)

# Load the initial page
url = "https://www.zara.com/kr/ko/man-the-best-sellers-l5899.html?v1=2351804"
driver.get(url)
time.sleep(5)  # Wait for page to fully load

# Accept cookies
try:
    accept_cookies_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="onetrust-accept-btn-handler"]')))
    accept_cookies_button.click()
except Exception as e:
    print("Failed to accept cookies:", e)

# Scroll down to load all images
body = driver.find_element(By.TAG_NAME, "body")
for _ in range(30):  # Adjust the number of scrolls as needed
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)  # Increase sleep time

# Parse the page source
soup = bs(driver.page_source, 'html.parser')
div_list = soup.find_all('div', attrs={'class': 'product-groups'})
div_data = div_list[0]
img_list = div_data.find_all('img')

# Counter for naming files
counter = 1

# Download and save images
for img in img_list:
    img_url = img['src']
    if 'transparent-background.png' not in img_url and 'loader.gif' not in img_url:
        # Handle URLs
        if img_url.startswith('//'):
            img_url = 'https:' + img_url
        
        # Extract filename from URL
        filename = f'zara_num{counter}.png'  # Adjust the extension if necessary
        
        # Download the image
        urllib.request.urlretrieve(img_url, f"./top/{filename}")
        
        print(f"{filename} saved")
        
        # Increment the counter
        counter += 1

driver.quit()
```
