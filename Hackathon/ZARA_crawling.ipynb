```python
import time
import os
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import requests
import urllib.request

import time
import os
import requests
import urllib.request
from selenium import webdriver
from bs4 import BeautifulSoup as bs
```

```python

driver = webdriver.Chrome()
url = f"https://www.zara.com/kr/ko/man-the-best-sellers-l5899.html?v1=2351804"
driver.get(url)
time.sleep(2)
```

```python
soup = bs(driver.page_source, 'html.parser')
```

```python
ul_list = soup.find_all('ul', attrs={
    'class' : 'product-grid__product-list'
})
```

```python
ul_data = ul_list[0]
ul_list[0]
```

```python
img_list = ul_data.find_all('img')

img_list
```

```python
import urllib.request

counter = 1

try:
    for img in img_list:
        img_url = img['src']
        
        # Handle URLs
        if img_url.startswith('//'):
            img_url = 'https:' + img_url
        
        # Get the filename with .png extension
        filename = f'ZARA_men{counter}.png'
        
        # Download the image
        urllib.request.urlretrieve(img_url, f"./ZARA_men/{filename}")
        
        print(f"{filename} saved")
        
        # Increment the counter
        counter += 1

    print("All images downloaded successfully.")

finally:
    # Close the web driver
    driver.quit()
```
