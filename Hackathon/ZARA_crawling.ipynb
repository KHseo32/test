```python
import time
import os
import requests
import urllib.request
from selenium import webdriver
from bs4 import BeautifulSoup as bs

# Create the 'top' directory if it doesn't exist
os.makedirs('./top', exist_ok=True)

driver = webdriver.Chrome()

# Iterate through pages 1 to 8
for page_num in range(1, 9):
    url = f"https://www.zara.com/kr/ko/man-the-best-sellers-l5899.html?v1=2351804"
    driver.get(url)
    time.sleep(2)
    soup = bs(driver.page_source, 'html.parser')
    ul_list = soup.find_all('div', attrs={'class' : 'product-groups'})
    ul_data = ul_list[0]
    img_list = ul_data.find_all('img')

    # Counter for naming files
    counter = 1

    for img in img_list:
        img_url = img['src']
        if img_url.endswith('.jpg'):
            # Handle URLs
            if img_url.startswith('//'):
                img_url = 'https:' + img_url
            
            # Get the filename with .png extension
            filename = f'ZARA_outers{counter}.png'
            
            # Download the image
            urllib.request.urlretrieve(img_url, f"./top/{filename}")
            
            print(f"{filename} saved")
            
            # Increment the counter
            counter += 1

driver.quit()
```
