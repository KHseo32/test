```python
def scrape_comments(url):
    # Send a GET request to the URL
    response = requests.get(url)
    # Parse the HTML content
    soup = bs(response.text, 'html.parser')
    # Find all comment elements
    comments = soup.find_all('div', class_='u_cbox_comment_box')
    
    # List to hold comment data
    comment_data = []
    
    # Extract data from each comment element
    for comment in comments:
        # Extract the comment text
        comment_text = comment.find('span', class_='u_cbox_contents').get_text(strip=True)
        # Extract the comment date
        comment_date = comment.find('span', class_='u_cbox_date').get_text(strip=True)
        # Extract the commenter's nickname
        commenter = comment.find('span', class_='u_cbox_nick').get_text(strip=True)
        
        # Append the data to the list
        comment_data.append({
            'Commenter': commenter,
            'Date': comment_date,
            'Comment': comment_text
        })
    
    # Return the list of comment data
    return comment_data
```

```python
import requests
from bs4 import BeautifulSoup
import os

# 웹 페이지 URL 설정
url = "https://www.musinsa.com/app/styles/lists?use_yn_360=&style_type=&brand=&tag_no=&display_cnt=60&list_kind=big&sort=NEWEST&page=1"

# 페이지 내용 가져오기
response = requests.get(url)

# BeautifulSoup으로 페이지 파싱
soup = BeautifulSoup(response.content, 'html.parser')

# div 태그 하위에 있는 img 태그의 src 속성 추출
thumbnails = soup.select("div img[src$='.png']")

# 이미지 저장 폴더 생성 (필요한 경우)
if not os.path.exists("thumbnails"):
    os.makedirs("thumbnails")

# 이미지 다운로드
for i, thumbnail in enumerate(thumbnails):
    src = thumbnail["src"]
    # 이미지 파일명 설정
    filename = f"thumbnails/thumbnail_{i+1}.png"
    # 이미지 다운로드
    with open(filename, "wb") as f:
        f.write(requests.get(src).content)
    print(f"다운로드 완료: {filename}")

print(f"총 {len(thumbnails)}개의 썸네일 이미지를 다운로드했습니다.")
```

```python
import time
import os
from selenium import webdriver
from bs4 import BeautifulSoup
import requests

# 웹 드라이버 로드
driver = webdriver.Chrome()

# 이미지 저장 폴더 생성 (필요한 경우)
if not os.path.exists("thumbnails"):
    os.makedirs("thumbnails")

for i in range(1, 131):  # Loop from page 1 to page 130
    # 웹 페이지 URL 설정
    url = f"https://www.musinsa.com/app/styles/lists?use_yn_360=&style_type=&brand=&tag_no=&display_cnt=60&list_kind=big&sort=NEWEST&page={i}"
    
    # 페이지 내용 가져오기
    driver.get(url)
    time.sleep(2)  # 페이지가 로드될 때까지 잠시 기다림
    
    # BeautifulSoup으로 페이지 파싱
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # div 태그 하위에 있는 img 태그의 src 속성 추출
    thumbnails = soup.find_all("img")
    
    # 이미지 다운로드
    for j, thumbnail in enumerate(thumbnails):
        src = thumbnail["src"]
        # URL에 scheme이 없는 경우 처리
        if not src.startswith("http"):
            src = "https:" + src
        # 이미지 파일명 설정
        filename = f"thumbnails/thumbnail_page_{i}_{j+1}.png"
        # 이미지 다운로드
        with open(filename, "wb") as f:
            f.write(requests.get(src).content)
        print(f"다운로드 완료: {filename}")
    
    print(f"페이지 {i}에서 {len(thumbnails)}개의 썸네일 이미지를 다운로드했습니다.")

# 웹 드라이버 종료
driver.quit()
```
