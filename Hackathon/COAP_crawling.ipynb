```python
## best


import time
import os
import requests
import urllib.request
from selenium import webdriver
from bs4 import BeautifulSoup as bs

# Create the 'COAP' directory if it doesn't exist
os.makedirs('./COAP', exist_ok=True)

driver = webdriver.Chrome()

# Iterate through pages 1 to 7
for page_num in range(1, 8):
    url = f"https://comeoutandplay.kr/product/list.html?cate_no=64&page={page_num}"
    driver.get(url)
    time.sleep(2)
    soup = bs(driver.page_source, 'html.parser')
    ul_list = soup.find_all('ul', attrs={'class': 'prdList grid3'})
    ul_data = ul_list[0]
    img_list = ul_data.find_all('img')

    # Counter for naming files
    counter = 1

    for img in img_list:
        img_url = img['src']
        if img_url.endswith('.jpg'):
            # Handle URLs
            if img_url.startswith('//'):
                img_url = 'https:' + img_url
            
            # Get the filename with .png extension
            filename = f'page{page_num}_COAP{counter}.png'
            
            # Download the image
            urllib.request.urlretrieve(img_url, f"./COAP/{filename}")
            
            print(f"{filename} saved")
            
            # Increment the counter
            counter += 1

driver.quit()
```
