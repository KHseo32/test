#### IFrame Popup

```python
from folium.map import Popup
import pandas as pd
import folium
df = pd.read_excel(path_Data_in+'서울지역 대학교 위치.xlsx')
seoul_map = folium.Map(location=[37.55,126.98], zoom_start=12)
```

```python
for name, lat, lng in zip(df.학교명, df.위도, df.경도):
    iframe = folium.IFrame(name, width=300, height=80)
    popup = folium.Popup(iframe, max_width = 300)
    folium.Marker([lat, lng], popup=popup).add_to(seoul_map)
seoul_map.save(path_Data_out+'seoul_colleges.html')
seoul_map
```

#### 원형 popup

```python
from folium.map import Popup
import pandas as pd
import folium
df = pd.read_excel(path_Data_in+'서울지역 대학교 위치.xlsx')
seoul_map = folium.Map(location=[37.55,126.98], zoom_start=12)
```

```python
for name, lat, lng in zip(df.학교명, df.위도, df.경도):
    folium.CircleMarker([lat, lng],
                       radious=10,
                       color='brown',
                       fill=True,
                       fill_color='coral',
                       fill_opacity=0.7,
                       popup=name
    ).add_to(seoul_map)
seoul_map.save(path_Data_out+'seoul_colleges2.html')
seoul_map
```

### 경기도 인구 변화

```python
import pandas as pd
import folium
import json

file_path = r'C:\Users\skh13\OneDrive\Documents\GitHub\ubion_5\project01\03. 데이터 분석\03. 시각화\Data-in\경기도인구데이터.xlsx'
df = pd.read_excel(file_path, index_col='구분', engine='openpyxl')
```

```python
df.columns = df.columns.map(str)
```

```python
geo_path = r'C:\Users\skh13\OneDrive\Documents\GitHub\ubion_5\project01\03. 데이터 분석\03. 시각화\Data-in\경기도행정구역경계.json'
try:
    geo_data = json.load(open(geo_path, encoding='utf-8'))
except:
    geo_data = json.load(open(geo_path, encoding='utf-8-sig'))
```

```python
## 경기도 지도 만들기
g_map = folium.Map(location=[37.5502, 126.982], tiles='Stamen Terrain',zoom_start=9)
```

```python

year = '2007'



folium.Choropleth(
    geo_data=geo_data,
    data=df[year],
    columns=[df.index, df[year]],  
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.3,
    threshold_scale=[10000, 100000, 300000, 500000, 700000],
    key_on='feature.properties.name'
    ).add_to(g_map)

# Save the map as an HTML file
g_map.save('./gyonggi_population_' + year + '.html')
```
