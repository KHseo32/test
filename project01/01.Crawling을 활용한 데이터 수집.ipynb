```python
# errorHttp
from urllib.request import urlopen
from urllib.request import HTTPError

try:
    html = urlopen("http://www.google.com/kim.html")
except HTTPError as e:
    print(e)
else:
    print("성공")
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
import time

service = Service()  # Selenium 서비스 생성
options = webdriver.ChromeOptions()  # Chrome 브라우저의 옵션을 설정
driver = webdriver.Chrome(service=service, options=options) # Chrome 브라우저 실행

sec = 2
url = "http://naver.com"

# driver는 Chrome을 제어할 수 있는 selenium임
driver.get(url)
time.sleep(sec)
driver.back()
time.sleep(sec)
driver.forward()
time.sleep(sec)
driver.refresh()
time.sleep(sec)
driver.back()
```

```python
# naver 로긴(우회해서 접속)
import pyperclip
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
url = "http://naver.com"
driver.get(url)
try:
    #네이버 이동
    driver.get(url)
    # 로그인 버튼 클릭
    element = driver.find_element(By.CLASS_NAME, "MyView-module__link_login___HpHMW")
    element.click()
    # id, pass 입력
    pyperclip.copy("아이디")
    driver.find_element(By.ID, "id").send_keys(Keys.CONTROL, 'v')
    pyperclip.copy("비밀번호")
    driver.find_element(By.ID, "pw").send_keys(Keys.CONTROL, 'v')
    driver.find_element(By.ID, "log.login").click()
    print(driver.page_source)
except Exception as e:
    print(e)
```

```python
productlist = driver.find_element(By.ID, 'productList')
lis=productlist.find_elements(By.TAG_NAME, 'li')

prd_name=[]
prd_price=[]
prd_deli=[]
prd_url =[]

for idx, li in enumerate(lis):
    try:
        prd_name.append(li.find_element(By.CLASS_NAME, 'name').text)
        prd_price.append(li.find_element(By.CLASS_NAME, 'price-value').text)
        prd_deli.append(li.find_element(By.CLASS_NAME, 'delivery').text)
        prd_url.append(li.find_element(By.CLASS_NAME, 'baby-product-link').get_attribute('href'))

        e = li.find_element(By.TAG_NAME, 'img')
        img_url = e.get_attribute('src')
        img_res = requests.get(img_url)
        img_res.raise_for_status()

        food_name = "./image/food{}.jpg".format(idx+1)
        with open(food_name, 'wb') as f:
            f.write(img_res.content)

        if idx >=5:
            break
    except Exception:
        pass



print(prd_deli)
print(prd_name)
print(prd_price)
print(prd_url)
# import pandas as pd

# # 데이터 프레임 생성
# df = pd.DataFrame({
#     'Product Name': prd_name,
#     'Price': prd_price,
#     'Delivery Details': prd_deli,
#     # 'Product URL': prd_url
# })

# # CSV 파일로 저장
# csv_filename = 'coupang_products.csv'
# df.to_csv(csv_filename, index=False)

# print(f"제품 정보가 '{csv_filename}' 파일로 성공적으로 추출되었습니다.")
```
