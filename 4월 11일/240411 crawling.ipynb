### bs4 라이브러리
- bs4 라이브러리 안에 있는 BeautifulSoup Class 사용
- html로 이루어진 문자형 데이터를 parsing 작업하여 데이터를 쉽게 추출하기 위한 Class
    - parsing : 데이터 타입 변경(Class 안에 데이터를 대입)
- html의 TAG를 기준으로 데이터 (contents)를 쉽게 추출
- 웹의 구조를 어느정도 파악하고 사용하게 되면 쉽게 접근 가능
- Parser를 활용해서 Pyhton에서 접근이 쉽게 객체(변수 + 함수)의 형태로 제공

```python
from bs4 import BeautifulSoup as bs
```

```python
# html로 이루어진 문자열을 bs Class에 대입 -> pasing
soup = bs(html_data, 'html.parser')
```

```python
soup
```

- 태그의 이름을 기준으로 검색(필터)
    - soup.태그명 : 해당 태그명의 첫 번째 태그를 출력
    - soup.태그명.string : 첫 번째 태그의 문자열(contents)을 출력
        ex) `<p>test<p>` --> test
    - soup.태그명['속성명'] : 첫 번째 태그의 해당 속성의 값을 출력
        ex) `<a href='https://www.google.com>구글</a>` --> 'https://www.google.com'

- find()
    - HTML 데이터에서 특정 태그의 첫 번째 정보를 출력
    - find(속성명 = 속성값) : 태거들 중 해당 속성명+ 첫번째 속성 값의 태그 정보 출력
    - 결과 데이터가 TAG의 형태로 출력
- find_all()
    - HTML 데이터에서 특정 태그의 모든 정보 출력
    - limit 매개변수 : 출력해주는 데이터의 개수 지정
    - 결과 데이터가 TAG들의 집합인 ResultSet 형태로 출력(문자열로 생각하면 list 데이터)

```python
html_data2 = response2.text

html_data2
```

```python
### BeautifulSoap을 이용한 parsing
soup2 = bs(html_data2,'html.parser')

soup2
```

```python
# div 태그 중 class가 section_sise_top인 태그의 수 확인
len(
    soup2.find_all('div', attrs={
        'class' : 'section_sise_top'
    })
)
```

```python
div_data = soup2.find('div', attrs={
    'class' : 'section_sise_top'
})

div_data
```

```python
# div_data에서 table 태그의 수 확인
len(
    div_data.find_all(
        'table'
    )
)
```

```python
## table 태그의 데이터들을 변수에 저장
tables = div_data.find_all('table')
```

```python
tables
```

```python
# tables이 첫번째 태그를 이용하여 데이터를 추출->df로 생성

_table = tables[0]
```

```python
type(_table)
```

```python
_table
```

```python
## table에서 thead 추출
head_data = _table.find('thead')

head_data
```

```python
th_data = head_data.find_all('th')
```

```python
_col = []
for tag in th_data:
    _col.append(tag.get_text())
```

```python

_col
```

```python
tbody_data = _table.find('tbody')

tbody_data
```

```python
## tbody_data에서 tr태그들을 모두 찾는다.
tr_data = tbody_data.find_all('tr')
```

```python
tr_data
```

```python
# tr_data에서 첫번째 원소를 추출하여 contents들을 리스트로 생성
_val = [tr_data[0].find('th').get_text()]

for tag2 in tr_data[0].find_all('td'):
    # print(tag2.get_text().strip())
    _val.append(tag2.get_text().strip())

_val
```

```python
## 위 코드 반복 실행하여 빈 리스트에 추가
_values = []

for i in range(len(tr_data)):
    _val = [tr_data[i].find('th').get_text()]
    for tag2 in tr_data[i].find_all('td'):
        _val.append(tag2.get_text().strip())
    _values.append(_val)

_values
```
