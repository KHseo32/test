### bs4 라이브러리
- bs4 라이브러리 안에 있는 BeautifulSoup Class 사용
- html로 이루어진 문자형 데이터를 parsing 작업하여 데이터를 쉽게 추출하기 위한 Class
    - parsing : 데이터 타입 변경(Class 안에 데이터를 대입)
- html의 TAG를 기준으로 데이터 (contents)를 쉽게 추출
- 웹의 구조를 어느정도 파악하고 사용하게 되면 쉽게 접근 가능
- Parser를 활용해서 Pyhton에서 접근이 쉽게 객체(변수 + 함수)의 형태로 제공

```python
# !pip install bs4
```

```python
from bs4 import BeautifulSoup as bs
```

```python
# html로 이루어진 문자열을 bs Class에 대입 -> pasing
soup = bs(html_data, 'html.parser')
```

```python
soup
```

- 태그의 이름을 기준으로 검색(필터)
    - soup.태그명 : 해당 태그명의 첫 번째 태그를 출력
    - soup.태그명.string : 첫 번째 태그의 문자열(contents)을 출력
        ex) `<p>test<p>` --> test
    - soup.태그명['속성명'] : 첫 번째 태그의 해당 속성의 값을 출력
        ex) `<a href='https://www.google.com>구글</a>` --> 'https://www.google.com'

- find()
    - HTML 데이터에서 특정 태그의 첫 번째 정보를 출력
    - find(속성명 = 속성값) : 태거들 중 해당 속성명+ 첫번째 속성 값의 태그 정보 출력
    - 결과 데이터가 TAG의 형태로 출력
- find_all()
    - HTML 데이터에서 특정 태그의 모든 정보 출력
    - limit 매개변수 : 출력해주는 데이터의 개수 지정
    - 결과 데이터가 TAG들의 집합인 ResultSet 형태로 출력(문자열로 생각하면 list 데이터)

```python
a_data = soup.find('a')
print(a_data)
print(a_data.string)
print(a_data.get_text())
```

```python
## resultSet 형태로 되어있는 데이터에서 contents 출력
a_data=soup.find_all('a')
print(a_data)
print(a_data[0].string)
print(a_data[1].get_text())
```

```python
result = []

for i in range(1, len(a_data), 1):
    result.append(a_data[i].get_text())

result
```

```python
result = []
i = 0

while i < len(a_data):
    result.append(a_data[i].get_text())
    i += 1

result
```

```python
result = list(map(lambda x: x.get_text(), a_data))


result
```
